```python
import os
import sys
sys.path.insert(0, os.path.abspath('/Users/sujitkhanna/Desktop/Courses/StrategyBetSizing'))

import matplotlib.pyplot as plt
import numpy
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from feature_modeling.feature_selection import *
from utils.bet_sizing import *
import plotly.graph_objects as go
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix

numpy.random.seed(0)
numpy.set_printoptions(suppress=True)

import warnings
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
%matplotlib inline
warnings.filterwarnings('ignore')

```

```python
strat_list = ["IDXOG","BO_SLOPE", "MULTI_PIVOT_BO", "VOL_CARRY"]
df_list = []
slippage = 0.0002

for strat in strat_list:
    file = f"data/meta_labels/{strat}_meta_labels.csv"
    df = pd.read_csv(file)
    df = df.loc[df["date"]>"2018-12-31"]
    df["date"] = pd.to_datetime(df["date"])
    df[f"{strat}_ret1"] = df["strat_ret"] - slippage*df["strat_trades"]
    df[f"{strat}_ret"] = df[f"{strat}_ret1"].cumsum()
    df = df.set_index(["date"])
    df_list.append(df[f"{strat}_ret"])
```

```python
df_new = pd.concat(df_list, axis=1)
df_new.head()
```

```python
def multi_line_plot(df_list, name_list, plot_name):
    fig = go.Figure()
    
    for i, df in enumerate(df_list):
        df = df.to_frame()
        fig.add_trace(go.Scatter(x=df.index, y=df[f"{name_list[i]}_ret"], name=name_list[i], mode='lines'))
    
    # Add figure title
    fig.update_layout(
        title_text=plot_name,
        template="plotly_dark"
    )

    # Set x-axis title
    fig.update_xaxes(title_text="date")

    # Set y-axes titles
    fig.update_yaxes(title_text="strategy_perf: cum returns")
    fig.show()
```

```python
multi_line_plot(df_list, ["IDXOG", "BO_SLOPE", "MULTI_PIVOT_BO", "VOL_CARRY"], "VXX Intraday Strategy Performance")
```
